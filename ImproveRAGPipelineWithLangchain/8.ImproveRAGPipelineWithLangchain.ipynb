````python
import re
from typing import List
from langchain_community.utilities.sql_database import SQLDatabase
from langchain.chains import create_sql_query_chain
from langchain_community.tools.sql_database.tool import QuerySQLDataBaseTool
from langchain_nvidia_ai_endpoints import ChatNVIDIA
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import json

class MultiQueryFixed:
    def __init__(self, question: str, db: SQLDatabase, llm):
        self.question = question
        self.db = db
        self.llm = llm

    # --- helper: clean sql text produced by write_query
    def extract_sql(self, raw: str) -> str:
        if not raw:
            return ""
        q = re.sub(r"(?i)sql\s*query\s*[:\-]*", "", raw)   # remove "SQLQuery:" prefix
        q = re.sub(r"```sql|```", "", q)                   # remove fences
        q = q.strip()
        return q

    # --- generate n paraphrases (multi-questions)
    def gen_multi_questions(self, n: int = 5) -> List[str]:
        template = (
            "Generate {n} alternative user queries (different wordings) "
            "for the following question. Return them as numbered lines:\n\n"
            "Original question: {question}\n\n"
            "Format exactly:\n1. ...\n2. ...\n...\n"
        )
        prompt = template.format(n=n, question=self.question)
        raw = self.llm.invoke(prompt)  # assume string result; adapt if response object
        raw_text = raw.content if hasattr(raw, "content") else str(raw)
        # split lines and keep those matching "^\d+\.\s+"
        lines = [line.strip() for line in raw_text.splitlines() if re.match(r"^\d+\.\s+", line.strip())]
        # remove numbering
        questions = [re.sub(r"^\d+\.\s*", "", l).strip() for l in lines]
        return [q for q in questions if q]

    # --- run SQL for each paraphrase
    def generate_and_execute(self, paraphrases: List[str]):
        write_query_chain = create_sql_query_chain(self.llm, self.db)   # factory chain
        execute_tool = QuerySQLDataBaseTool(db=self.db)
        all_results = []
        for q in paraphrases:
            # 1) generate SQL
            raw_sql = write_query_chain.invoke({"question": q})
            sql = self.extract_sql(raw_sql)
            if not sql:
                continue
            # optionally remove LIMIT if user asked ALL
            if re.search(r"\b(all|every|as many as possible)\b", self.question, re.I):
                sql = re.sub(r"(?i)\bLIMIT\s+\d+\b", "", sql)
            # 2) execute safely
            try:
                result = execute_tool.invoke(sql)
            except Exception as e:
                result = []
            # result is typically list of tuples; store with metadata
            all_results.append({
                "paraphrase": q,
                "sql": sql,
                "rows": result
            })
        return all_results

    # --- flatten + unique preserve order, convert rows to strings
    def unique_rows(self, list_of_results):
        seen = set()
        out = []
        for entry in list_of_results:
            rows = entry.get("rows") or []
            for row in rows:
                # join multi-col row into single string
                if isinstance(row, (list, tuple)):
                    s = " | ".join([str(c) for c in row if c is not None])
                else:
                    s = str(row)
                if s not in seen:
                    seen.add(s)
                    out.append(s)
        return out

    # --- final answer generation (use all rows or JSON)
    def answer_from_rows(self, unique_rows_list: List[str]):
        # prepare result_text. If many rows, keep all or limit per policy.
        # Use a strict prompt that forbids inventing rows.
        result_text = "\n".join(unique_rows_list)
        answer_prompt = ChatPromptTemplate.from_template(
            "You are an AI assistant. Use ONLY the SQL result below. Do NOT invent rows.\n\n"
            "Question: {question}\n\n"
            "SQL Result (each row on its own line):\n{result_text}\n\n"
            "Return a short, exact answer (list all rows)."
        )
        prompt = answer_prompt.format_prompt(question=self.question, result_text=result_text)
        response = self.llm.invoke(prompt.to_string())
        return response.content if hasattr(response, "content") else str(response)

    # --- main run
    def run(self):
        paraphrases = self.gen_multi_questions()
        if not paraphrases:
            return "No paraphrases generated."
        results = self.generate_and_execute(paraphrases)
        unique = self.unique_rows(results)
        # if user explicitly asked ALL, return programmatically (no LLM) to avoid hallucination
        if re.search(r"\b(all|every|as many as possible)\b", self.question, re.I):
            return unique
        # else call LLM to format answer
        return self.answer_from_rows(unique)
````

````python
LLM_MODEL = "nvidia/nvidia-nemotron-nano-9b-v2"
db_name = "Finance_data.db"
db_path = f"sqlite:///{db_name}"

llm = ChatNVIDIA(model=LLM_MODEL)
db = SQLDatabase.from_uri(
        db_path,
        include_tables=[
            "AsamaMove_Sales",
            "Supermarket_Sales",
            "Employee_Sales",
        ]
)
````

````python
query = "Show the total number of customers in each city, and the city with the highest number."
multi_query = MultiQueryFixed(question=query, db=db, llm=llm)
multi_query.run()
````
